```python
from deepface import DeepFace
import matplotlib.pyplot as plt
```

```python
sample = '../data/wiki_crop/00/37500_1944-01-23_2010.jpg'
```

```python
backends = ['opencv', 'ssd', 'dlib', 'mtcnn']
detected_face = DeepFace.detectFace(sample, detector_backend = backends[2])
plt.imshow(detected_face)
```

```python
aligned_face = DeepFace.analyze(sample, detector_backend = backends[2])
plt.imshow(aligned_face)
```
